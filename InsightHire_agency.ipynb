### Why find and find_all doesn't work for derivateve variables

###### Remove the '#' from the code below and read it to understand

```python
# # Extracting the job description from the first job data in the jobs variable we extracted before

# job_description1 = jobs.find('li', class_ = 'job-desc')
# job_description = job_description1.text.strip() if job_description1 else ''
# print(job_description)

# # You can't use the find method or find_all method on a derivative variable of beautiful soup. Because beautifulsoup returns a list, they expect you to just loop through so they didn't code the find and find_all function to work on a derivative of the original request data.
# # The only way to use find or find_all on a derivative variable is if you indicated the index in the list, since beautifulsoup returns a list. so the above example becomes...Jobs[0] or jobs[1] or jobs[2] etc, as seen below...
job_description1 = jobs[0].find('li', class_ = 'job-desc')
job_description = job_description1.text.strip() if job_description1 else ''
job_description
```

#### Some other ways to use find and find_all with ''='' and '':'' or dictionary version

```python
# Extracting the job deadline...
job_deadline = jobs[0].find('li', id = 'job-date').text
print(job_deadline)

#or

job_deadline = jobs[0].find('li', class_ = 'job-item').text
print(job_deadline)

#or

job_deadline = jobs[0].find('li', {'id' : 'job-date'}).text
print(job_deadline)


# If you check through, you'll notice when using the '=' version, it includes and underscors(_) when it's the class attribute but if its a different attribute, you'll not add the underscore(_)
```

### Creating a dataframe from the available data so far

```python
# First is to create an empty dictionary
jobs_dict = {
    'job_name' : [],
    'company_name' : [],
    'job_description' : [],
    'job_deadline' : []
}

web_pages = range(1, 4)
for web_page in web_pages:
    url = f'https://www.myjobmag.com/page/{web_page}'
    response = requests.get(url)

    soup = BeautifulSoup(response.text, 'html.parser')

    # Next is to loop through the requested data but this time since its a derived variable data, you'll follow the guide i layed down above, in this case, we'll just go back and copy exactly what we did in the previous cells
    for job in jobs:
        job_title = job.find('li', class_ = 'mag-b')
        job_description1 = job.find('li', class_ = 'job-desc')
        job_deadline1 = job.find('li', id = 'job-date')

        if job_title:
            if 'at' in job_title.text:
                job_title = job_title.text.split('at')
            
                job_name = job_title[0].strip()
                company_name = job_title[1].strip()
            
            else:
                print(f"no 'at' can be found in {job_title}")
        
        else:
            print(f'no value for {job_title}')

        
        job_description = job_description1.text.strip() if job_description1 else print(f'no {job_description1}')
        job_deadline = job_deadline1.text if job_deadline1 else print(f'no {job_deadline1}')

        jobs_dict['job_name'].append(job_name)
        jobs_dict['company_name'].append(company_name)
        jobs_dict['job_description'].append(job_description)
        jobs_dict['job_deadline'].append(job_deadline)



print(job_deadline)




##Here's a more roboust version of this code...

# import requests
# from bs4 import BeautifulSoup

# # Create an empty dictionary to store job data
# jobs_dict = {
#     'job_name': [],
#     'company_name': [],
#     'job_description': [],
#     'job_deadline': []
# }

# web_pages = range(1, 4)

# # Loop through the pages
# for web_page in web_pages:
#     url = f'https://www.myjobmag.com/page/{web_page}'
#     try:
#         response = requests.get(url)
#         response.raise_for_status()  # Raise an error for bad responses (4xx, 5xx)
#     except requests.RequestException as e:
#         print(f"Error fetching page {web_page}: {e}")
#         continue  # Skip this page if an error occurs

#     soup = BeautifulSoup(response.text, 'html.parser')

#     # Locate the job listings properly
#     jobs = soup.find_all('div', class_='job-list')  # Adjust class based on actual structure
#     if not jobs:
#         print(f"No jobs found on page {web_page}")
#         continue

#     # Loop through the jobs extracted
#     for job in jobs:
#         try:
#             job_title = job.find('li', class_='mag-b')
#             job_description1 = job.find('li', class_='job-desc')
#             job_deadline1 = job.find('li', id='job-date')

#             # Title extraction and validation
#             if job_title and 'at' in job_title.text:
#                 job_title_parts = job_title.text.split('at')
#                 job_name = job_title_parts[0].strip()
#                 company_name = job_title_parts[1].strip()
#             else:
#                 job_name = job_title.text.strip() if job_title else None
#                 company_name = None
#                 print(f"Warning: Could not extract job title properly on page {web_page}")

#             # Description and deadline extraction
#             job_description = job_description1.text.strip() if job_description1 else None
#             job_deadline = job_deadline1.text.strip() if job_deadline1 else None

#             # Append data to the dictionary only if valid
#             jobs_dict['job_name'].append(job_name)
#             jobs_dict['company_name'].append(company_name)
#             jobs_dict['job_description'].append(job_description)
#             jobs_dict['job_deadline'].append(job_deadline)

#         except AttributeError as e:
#             print(f"Error parsing job data on page {web_page}: {e}")
#             continue  # Skip this job if data parsing fails

# # Display the collected job deadlines
# print("Job Deadlines Collected:", jobs_dict['job_deadline'])

```

### Cleaning the date further


```python
def parse_date(date_str):
    if pd.isna(date_str) or not isinstance(date_str, str):
        return None
    
    if len(date_str.split()) == 2:
        date_str = f'{date_str} {datetime.now().year}'
        return datetime.strptime(date_str, '%d %B %Y').date()
    return datetime.strptime(date_str, '%d %B %Y').date()

jobs_df['job_deadline'] = jobs_df['job_deadline'].apply(parse_date)



# def parse_date(date_str):
#     # Check for None or invalid types
#     if pd.isna(date_str) or not isinstance(date_str, str):
#         return None  # Return None for missing or invalid data

#     # If only day and month are provided, add the current year
#     elif len(date_str.split()) == 2:
#         date_str = f"{date_str} {datetime.now().year}"
    
#     # Try to parse the date
#     try:
#         return datetime.strptime(date_str, '%d %B %Y').date()
#     except ValueError as e:
#         print(f"Error parsing date: {date_str} - {e}")
#         return None  # Return None for invalid formats

# # Apply the corrected function to your DataFrame
# jobs_df['job_deadline'] = jobs_df['job_deadline'].apply(parse_date)
```

### Define database connection parameters

```python
# Load environment variables from the .env file
load_dotenv()

db_params = {
    'username' : os.getenv('DB_USER'),
    'password' : os.getenv('PASSWORD'),
    'host' : os.getenv('HOST'),
    'port': int(os.getenv('PORT')),  # Ensure PORT is an integer
    'database' : os.getenv('DATABASE')
}

# define the database connection url
db_url = f"postgresql://{db_params['username']}:{db_params['password']}@{db_params['host']}:{db_params['port']}/{db_params['database']}"


# create the database engine with the db url specified
engine = create_engine(db_url)

# connect to postgreSQL server
with engine.connect() as connection:
    # Create table and load the data
    jobs_df.to_sql('InsigntHire_agency_jobs', connection, index = False, if_exists = 'replace')

print('Database, table and data loaded successfully')


```
